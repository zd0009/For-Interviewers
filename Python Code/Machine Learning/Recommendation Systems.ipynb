#### Create a item based similarity score matrix

```python
rs_ibs = pd.DataFrame(index=rs.columns,columns=rs.columns)
```

```python
t1_start = process_time()  

for i in range(len(rs.columns)):
    for j in range(len(rs.columns)):
         rs_ibs.ix[i,j] = 1-cosine(rs.ix[:,i],rs.ix[:,j])
t1_stop = process_time() 
t1_stop-t1_start
```

```python
rs_ibs.fillna(0, inplace = True)
rs_ibs
```

### 1. Name 2 other similarity measures that you can use instead of cosine similarity above.

***1. Euclidean distance can used as similarity measure for recommendation systems. For two vectors q and p, the Euclidean distance is calculated by:*** <br>
    $d(p,q)= \sqrt{(p_{1}-q_{1})^{2} + (p_{2}-q_{2})^{2}+...+(p_{n}-q_{n})^{2}}$ <br>
    
***2. Jaccard Similarity is another similarity measure. It is given by:***
    $s_{ij}=p/(p+q+r)$ <br>
 ***Where:<br>
 p = the number of attributes that are 1 for both objects. <br>
 q = Number of attributes 1 for i and 0 for j. <br>
 r = Number of attributes 0 for i and 1 for j.<br>***

### 2.  What is needed to build a Content-Based Recommender system?

***On top of the user and item profiles, a Utility Matrix is needed for Content-Based Recommender system. The ultility Matrix signifies content characteristics, i.e. the user’s preference with certain items. In the data gathered from the user, we have to find some relation between the items which are liked by the user and those which are disliked, for this purpose we use the utility matrix. In it we assign a particular value to each user-item pair, this value is known as the content characteristics or degree of preference. Then we draw a matrix of a user with the respective items to identify their preference relationship. We then match a user to the content characteristics in Utility Matrix as opposed to other users.***

### 3. Name 2 methods to evaluate your recommender system.

***1. Precision and Recall at Cutoff k is an evaluation metric for a recommender systems. Precision and Recall at cutoff k, P(k) and r(k), are simply the precision and recall calculated by considering only the subset of your recommendations from rank 1 through k. P(k) is the percentage in the top-k items that our recommender system recommends that a user is really interested in. R(k) is the percentage in all those items that a user is interested in that are covered by the recommeder system. A problem with recall and precision is that usually, recall increases as k is increased while precision decreases as k is increased.  Therefore, the F1 metric combines both measures: F1 = (2 * recall * precision)   /   (recall + precision). Computing F1 for each user and then taking the average is a good evaluation metric for top-k recommendation list.***


***2. Mean Average Precision at K (MAP@K) is another metric of choice for evaluating the performance of a recommender systems. If you have an algorithm that is returning a ranked ordering of items, each item is either hit or miss (like relevant vs. irrelevant search results) and items further down in the list are less likely to be used (like search results at the bottom of the page), MAP@K can be a good choice. MAP@K gives insight into how relevant the list of recommended items are.*** <br>

***The formular of MAP@K is:*** <br>

$MAP@K = \frac{1}{|U|} \sum_{u=1}^|U| \frac{1}{m}\sum_{k=1}^K P_u(k)\cdot rel_u(k).$ <br>

***Where U is the number of users, m is the number of revelant items and K is the the number of items to be recommended. where rel(k) is just an indicator that says whether that kth item was relevant (rel(k)=1) or not (rel(k)=0). P(k) is the precision calculated by considering only the subset of your recommendations from kth item.***


## 3. Challenge question - extra points (20)

Use a combination of user-item approach to build a recommendation score for each song for each user using the following steps for each user- <br>

1. For each song for the user row, get the top 10 similar songs and their similarity score. <br>
2. For each of the top 10 similar songs, get a list of the user purchases. <br>
3. Calculate a recommendation score as follows: ∑(purchaseHistory⋅similarityScore)∑similarityScore. <br>
4. What are the top 5 song recommendations for user 1606?.

##### For step 1

We have already created a item based similarity score matrix in Part 1A. It is called rs_ibs.<br>

We also created a matrix with top 10 song names recommended for each song. It is called rs_t10.

##### Step 2 and 3 <br>
For user U: <br>
for each song S: <br>
Find top 10 cloest songs to S and each similarity Score (similarityScore): <br>
Out of those top 10 closest songs, find the ones that have been listened to by user U (purchaseHistory)<br>
Calculate a recommendation score as follows: ∑(purchaseHistory⋅similarityScore)∑similarityScore. <br>
use the above 2 info to get the recommendation score for song S for user U <br>

```python
def iub_score(user, song):
 # Note that we score items that the user has already consumed as 0, because there is no point recommending it again.
    if rs.loc[user,song] == 1:
        rs_uib_score= 0
    else:
        # Here we find the top 10 songs names recommended for each song S from Part 1A's rs_t10
        song_top_names = rs_t10.loc[song,1:10]
         # Then we find the top 10 songs similarity scores for each song S from Part 1A's rs_ibs
        song_top_sims = rs_ibs.loc[song,:].sort_values(ascending=False)[1:11]
        # Out of those top 10 closest songs, find the ones that have been listened to by each user U 
        user_lis =rs.loc[user,song_top_names]
         # calculate similarity score for each user and each song
        rs_uib_score = sum(user_lis*song_top_sims)/song_top_sims.sum()
    return rs_uib_score
```

```python
# given a user name and song name we can calculate a user-item based score
iub_score(1606, 'david bowie')
```

##### What are the top 5 song recommendations for user 1606?.

```python
def top5_songs(user):
    scores = [iub_score(user, song) for song in rs.columns]
    top5 = pd.Series(scores, index = rs.columns).dropna().sort_values(ascending = False).head()
    return top5
```
