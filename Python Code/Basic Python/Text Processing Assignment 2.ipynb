```python
#Loading Libraries
import pandas as pd
from collections import Counter
from collections import defaultdict
import numpy as np
import random
```

### 1. Clean your input text by removing all punctuation, except periods. Convert everything to lower case.

```python
with open('walden.txt', encoding='utf-8') as waldenf:
    walden = waldenf.read().lower().replace(':','').replace(';','').replace(',','').replace('\n\n','\n').replace('\n',' ')
    walden = walden.replace('.', '.\n').replace('?', '.\n').replace('\n ','\n').replace(' \n','\n')
    walden = walden.replace("“",'').replace("”",'').replace("’",'').replace("————",'').replace("*",'')
    walden = walden.replace('"','').replace('!','').replace("'",'').replace('(','').replace(')','').replace('_','')
    walden = walden.replace('[','').replace(']','')
print(walden)
```

### 2. For each sentence, create pairs of consecutive words

```python
# replacing . with ' PERIOD'  
walden=walden.replace('.',' PERIOD')
print(walden)
```

```python
# Splitting the text to a list of words
mylist=walden.split()
mylist
```

```python
# Creating the paired-words list
paired_words=list(zip(mylist,mylist[1:]))
paired_words
```

### 3. Count how often each pair of words occurs in the input text and record it for later retrieval.

#### Dictionary

```python
# creating a pair_counter dictionary
pair_counter = dict()
for pairs in paired_words:
    w1 = pairs[0]
    w2 = pairs[1]    
    if w1 in pair_counter:
        if w2 in pair_counter[w1]:
            pair_counter[w1][w2] += 1
        else:
            pair_counter[w1][w2] = 1
    else:
        pair_counter[w1] = {w2:1}        
pair_counter        
```

#### Pandas

```python
# creating a list for my col and row names of the data frame
name=sorted(list(set(mylist)))
# removing 'PERIOD', numbers and symbols to the end of the list
name1=name[(name.index('PERIOD')+1):]
name1.extend(name[:(name.index('PERIOD')+1)])
print(name1[:10])
print(name1[-10:])
```

```python
#create dataframe and assign zeros to the frame
word_matrix=pd.DataFrame(index=name1, columns=name1)
for col in word_matrix.columns:
    word_matrix[col].values[:] = 0
word_matrix
```

```python
# assign counting values to the word matrix
for pair in paired_words:
    word_matrix[pair[0]][pair[1]] += 1
word_matrix    
```

### 4. Normalize the counts and turn them into probabilities.

#### Dictionary

```python
for w1 in pair_counter:
    totc = sum( pair_counter[w1][w2] for w2 in pair_counter[w1])   
    for w2 in pair_counter[w1]:
          pair_counter[w1][w2] /= totc
pair_counter
```

#### Pandas

```python
for ide in word_matrix.index:
    ss=sum(word_matrix.loc[ide,:])
    if ss !=0:
        word_matrix.loc[ide,:]=word_matrix.loc[ide,:]/ss
word_matrix    
```

### 5. Generate text based on this transition matrix

```python
# 100 random start words
startwords = list(np.random.choice(list(word_matrix.index), size=100))
startwords
```

#### Dictionary

```python
def sentend(start_word, pair_counter):
    sentence = start_word
    next_word=start_word
    for i in range(len(pair_counter)):
        if next_word != 'PERIOD':
            next_word = random.choices(list(pair_counter[next_word].keys()), weights = list(pair_counter[next_word].values()))[0]
            sentence += ' ' + next_word
        else: break
    return sentence
```

```python
sentencesd = []
for stwordd in startwords:
    sentencesd.append(sentend(stwordd, pair_counter))
```

#### Pandas

```python
def senten(start_word, word_matrix):
    sentence = start_word
    next_word=start_word
    for i in range(len(word_matrix.columns)):
        if next_word != 'PERIOD':
            next_word = random.choices(word_matrix.columns, weights = word_matrix.loc[next_word,:])[0]
            sentence += ' ' + next_word
        else: break
    return sentence
```

```python
sentences = []
for stword in startwords:
    sentences.append(senten(stword, word_matrix))
```
