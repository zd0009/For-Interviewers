#### 3. Print the first 30 lines, after you have: 
A. Removed commas, periods, colons and semicolons and  <br />
B. Converted all characters to lower case and  <br />
C. Removed any extra spaces (hint strip())  <br />

```python
for i in range(len(List_lines)):
    # remove commas, periods, colons and semicolons
    List_lines[i] = List_lines[i].translate(str.maketrans("","", ",.:;"))
    #Converted all characters to lower case and
    List_lines[i] = List_lines[i].lower()
    #  Removed any extra spaces (hint strip())
    List_lines[i] = List_lines[i].strip()
# remove space lines
List_lines_cleaned = []
for st in List_lines:
    if (st != ""):
        List_lines_cleaned.append(st)
    
for i in range(30):
    print(List_lines_cleaned[i])
```

#### 4. Count the number of unique words in the file (after transforming the text using logic from #3). How often are the following words used: "dawn", "distasteful", "employment?"

```python
# allocate a word list
List_words=[]
# append all the words to the word list
for element in List_lines_cleaned:
    List_words.extend(element.split(" "))
# I am creating a list of unique enlish words which only contain letter and hyphen -. Note '-' along is not a word
Eng_words=[]

for word in List_words:
    if bool(re.match("^[a-z-]*$", word)) and word not in Eng_words and word != '' and word != '-':
           Eng_words.append(word)

print("Number of unqiue words in the file: ", len(Eng_words))
print("Number of times 'dawn' used: ",List_words.count('dawn'))
print("Number of times 'distasteful' used: ", List_words.count('distasteful'))
print("Number of times 'employment' used: ", List_words.count('employment'))
```

#### 9. Create a file containing a "cleaned" version of the text with the following transformations: 
A. remove extra spaces and reformat the text so there is one English sentence per line. (this is done most easily if you load the whole file, rather than processing it line by line)  <br />
B. remove all punctuation  <br />
C. convert all characters to lower case  <br />


```python
# Create a new line list and only apply strip
file = open(data_file_location, 'r', encoding='utf8')
List_lines_strip =[]
for line in file:
    List_lines_strip.append(line.replace('\n', ' ').strip())
file.close()
# concat all the lines in to a string
all_text = ''.join(List_lines_strip)
# Sentences are seperated by period. Note ther is not period in the header except for url
sent_list=all_text.split('.')
# The first 4 sentences are seperated by periods in the url, so I am removing first 4 sentences
sent_list=sent_list[4:]
# remove all empty lines
sent_list_cleaned = []
for it in sent_list:
    if (it != ""):
        sent_list_cleaned.append(it)

# adding \n back to end of each line 
sent_list_cleaned = [item + '\n' for item in sent_list_cleaned]
#remove all punctuation and convert all characters to lower case
sent_list_cleaned = [item.translate(str.maketrans("","", string.punctuation+"“"+"”"+"’"+"————")).lower() for item in sent_list_cleaned]
# re-concatenate all cleaned lines back to a big string
all_text_clean=''.join(sent_list_cleaned)
text_file = open("cleaned_walden.txt", 'w', encoding='utf8')
text_file.write(all_text_clean)
text_file.close()
```
