```python
import csv
import requests
from bs4 import BeautifulSoup
from IPython.display import HTML
import pandas as pd
```

```python
# Scraping table data from following site
url = 'http://www.chicagocomputerclasses.com/average-city-chicago-income/'
headers = {'User-Agent': "Chrome/54.0.2840.90"}
response = requests.get(url, headers=headers)
html = response.content 
```

```python
HTML('<iframe src=http://www.chicagocomputerclasses.com/average-city-chicago-income/ width=900 height=350></iframe>')
```

```python
soup= BeautifulSoup(response.text, "lxml")
table = soup.find('table', attrs={'width':'283', 'border':'0', 'cellspacing':'0', 'cellpadding':'0'})
```

```python
rows = table.find_all('tr')
column_names = [v.text.replace('Average','Average of PER CAPITA INCOME') for v in rows[2].find_all('th')]
```

```python
df = pd.DataFrame(columns=column_names)

for i in range(3,len(rows)):
    tds = rows[i].find_all('td')
    values = [td.text for td in tds]
    df = df.append(pd.Series(values, index=column_names), ignore_index=True)
```
