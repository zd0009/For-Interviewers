```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('sports_need_update.csv')
df.head()
```

```python
df.shape
```

```python
df1= pd.DataFrame()
df1['Date']=df['Date'].unique()
df1['Bulls_28']=0
df1['Bears_33']=0
df1['Cubs_06']=0
df1['WhiteSox_34']=0
df1
```

```python
for i in range(len(df)):
    for j in range(len(df1)):
        if df1.Date[j] == df.Date[i]:
            if df.Team[i] == 'Bulls':
                df1['Bulls_28'][j] = 1
            elif df.Team[i] == 'Bears':
                df1['Bears_33'][j] =1
            elif df.Team[i] == 'Cubs':
                df1['Cubs_06'][j] =1
            elif df.Team[i] =='WhiteSox':  
                df1['WhiteSox_34'][j] =1
```
